```python
def reshape_dataframe(df):
    first_timestamps = df.groupby('id')['timestamp'].first().reset_index()

    pivot_df = df.pivot(index=['id', 'nominal_power', 'location_id'],
                        columns='sub_id',
                        values=['soc', 'power']).reset_index()

    pivot_df.columns = [
        f'{col[0]}_{col[1]}' if col[1] != '' else col[0]
        for col in pivot_df.columns
    ]

    result_df = pivot_df.merge(first_timestamps, on='id')
    return result_df
```

```python
train_data_full = reshape_dataframe(train_data_full)
```

```python
train_data_full.head()
```

```python
train, temp = train_test_split(train_data_full, train_size=0.70, test_size=0.30, shuffle=True, random_state=42)
validation, test = train_test_split(temp, train_size=0.5, test_size=0.5, random_state=42)
```

```python
# Define input features and target columns
INPUT_FEATURES = [f'power_{i}' for i in range(10)] # + [f'soc_{i}' for i in range(10)]
INPUT_FEATURES += ["nominal_power","location_id"]
# todo location_id should be discrete
# todo add timestamp
TARGETS = [f'power_{i}' for i in range(10, 40)]
```

```python
train = train.dropna()
validation = validation.dropna()
test = test.dropna()
```

```python
# Prepare training data
X_train = train[INPUT_FEATURES]
y_train = train[TARGETS]
```

```python
# Prepare validation data
X_validation = validation[INPUT_FEATURES]
y_validation_true = validation[TARGETS]

# Prepare test data
X_test = test[INPUT_FEATURES]
y_test = test[TARGETS]
```

```python
X_train.head()
```

```python
pipeline = Pipeline([
    ('scaler', MinMaxScaler()),
    ('poly', PolynomialFeatures()),
    ('regressor', MultiOutputRegressor(Ridge()))
])

param_distributions = {
    'poly__degree': [1, 2, 3],
    'regressor__estimator__alpha': [0.1,1,10],
    'scaler': [MinMaxScaler(), StandardScaler()]
}

grid_search = RandomizedSearchCV(
    estimator=pipeline,
    param_distributions=param_distributions,
    n_iter=6,  # Number of parameter settings that are sampled
    cv=3,
    scoring='neg_mean_squared_error',
    n_jobs=1,
    verbose=2,
    random_state=42,
)
```

```python
# Fit the model
grid_search.fit(X_train, y_train)
```

```python
print("Best parameters found:")
print(grid_search.best_params_)

# Best estimator
multi_output_model = grid_search.best_estimator_
```

```python
multi_output_model.fit(X_train, y_train)
y_validation_pred = multi_output_model.predict(X_validation)
y_test_pred = multi_output_model.predict(X_test)
```

```python
y_validation_pred = multi_output_model.predict(X_validation)
y_test_pred = multi_output_model.predict(X_test)
```

```python
# Evaluate performance on validation data
mse_validation = mean_squared_error(y_validation_true, y_validation_pred)
mae_validation = mean_absolute_error(y_validation_true, y_validation_pred)

print(f'Validation MSE: {mse_validation}')
print(f'Validation MAE: {mae_validation}')
```

```python
def plot_graphs(index_start, index_end):
    num_samples = index_end - index_start + 1
    fig, axs = plt.subplots(num_samples, 1, figsize=(12, 4 * num_samples))
    
    # If there's only one sample, axs may not be a list; make it a list for consistency
    if num_samples == 1:
        axs = [axs]
    
    for idx, sample_index in enumerate(range(index_start, index_end + 1)):
        # Get the true 'power' values from time steps 0 to 39
        true_values = validation.iloc[sample_index][[f'power_{i}' for i in range(40)]].values
        
        # Get the known 'power' values from time steps 0 to 9
        known_power_values = validation.iloc[sample_index][[f'power_{i}' for i in range(10)]].values
        
        # Get the predicted 'power' values from time steps 10 to 39
        predicted_values_future = y_validation_pred[sample_index]
        
        # Combine the known 'power' values with the predicted values
        predicted_values = np.concatenate([known_power_values, predicted_values_future])
        
        # Plotting on the subplot
        axs[idx].plot(range(40), true_values, label='True Values')
        axs[idx].plot(range(40), predicted_values, linestyle='--', label='Predicted Values')
        axs[idx].set_xlabel('Time Step')
        axs[idx].set_ylabel('Power')
        axs[idx].set_title(f'Predicted vs True Power Values (Sample {sample_index})')
        axs[idx].legend()
        axs[idx].grid(True)
    
    plt.tight_layout()
    plt.show()
```

```python
plot_graphs(100,110)
```
